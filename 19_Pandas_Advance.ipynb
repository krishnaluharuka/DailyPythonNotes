# Accessing Row and Column of DataFrame

## Accessing 1st nth row

`dataframe[x:y:z]`

- x : Start , y : End, z = Steps
- Just like string slicing

```python
df
```

```python
df[0:100]
```

```python
df[0:100:5]
```

## Implicit / Integer / Internal Indexing

iloc = implicit , integer or internal indexing

- Always require python internal indexing
- whatever indexing Pandas give by-default , we can use it with 

```python
df[0:3]
```

```python
df.iloc[0:3]  # start from 0 and go till 2 (3-1)
```

## Explicit / Name Indexing

These indexing in which end point is also included.

```python
df.loc[0:3]
```

```python
df.loc[0:2, ["Name", "Sex"]]  # if we use iloc in place of loc, we will get error as iloc in number indexing
```

If we want to use iloc only, we have to use indexing in place of name.
- like in place of col name "Sex" we will use index no. 4

```python
df.iloc[0:3, 3:5]
```

### Dropping any column from Data Frame

Whatever keys we are dropping, by default it is assuming in the row.
- We have to drop by giving parameter axis=1

- By default not modified ni original data frame.
- In order to modify from original data frame, we pass a parameter inplace= True

```python
# df.drop('PassengerId') # error
```

```python
df.drop("PassengerId", axis=1)
```

```python
df
```

```python
df.drop("PassengerId", axis=1, inplace=True)
```

```python
df
```

Dropping the row, just mention the index of the row to drop, and inplace=True if want to drop from original data frame

```python
df.drop(1, inplace=True) 
```

```python
df
```

```python
df.drop(4, inplace=True)
```

```python
df
```

To see original data frame after dropping

```python
df.reset_index(drop=True)
```
