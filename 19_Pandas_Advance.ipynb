- Numerical data will be replaced by mean, median.
- Categorical Data will be replaced by Mode.

### Imputation 

Replacing missing value with some value.
- Numeric data will be replaced by mean and median >>> Outlier(abnormal) value treatment is done , then use mean. otherwise median

- Categorical data will be replaced by Mode.
- Input missing value with someconstan

```python
df1.fillna("SomeValue")
```

Since parent_id is numerical col, we can replace NAN by 0 as

```python
df1.fillna(0)
```

```python
data1 = {
    'A': [1, 2, None, 4, 5, None, 7, 8, 9, 10],
    'B': [None, 11, 12, 13, None, 15, 16, None, 18, 19]
}
df2 = pd.DataFrame(data1)
```

```python
df2
```

Calculating Mean of Col

```python
df2['A'].mean()
```

Missing value is imputed by mean oc 

```python
df2['A'].fillna(df2['A'].mean())
```

```python
df2['A'].median()
```

```python
df2['A'].fillna(df2['A'].median())
```

```python
df2.fillna(0)
```

### Other way to impute missing value

`f-fill` : Forward filling

See the data from the bottom and fill the previous observed value to NaN value

```python
df2.ffill()
```

`bfill` : Backward Filling

See the data frame from top and fill the next observed value

```python
df2.bfill()
```
