# Introduction on Seaborn

**Statistical Data Visualization**

Seaborn is a Python data visualization library built on top of Matplotlib.

It comes with prettier default styles and is optimized for statistical plots.

Makes it easier to create informative, attractive, and complex plots with less code.

## Uses 

- Cleaner, modern designs without manually styling.

- Built-in statistical functions (like regression lines, distribution plots).

- Works directly with Pandas DataFrames (no need to manually extract lists).

- Supports themes, color palettes, and automatic legends.

- Offers specialized plots that Matplotlib doesn’t have natively.

## Scatter Plot

How tip is varying wrt total bill?

```python
sns.relplot(x= 'total_bill', y= 'tip', data= df)
```

*Insights* : As the bill increases, Tips also increases >> @ $50 tip is $10

Relationship between two variable,  By default it is scatter plot, kind = line if want to change the line

```python
df
```

To distinguish between smoker and non smoker

```python
sns.relplot(x= 'total_bill', y= 'tip', data= df, hue= 'smoker')
```

*Insights* : 
- Number of smokers are comparatively more in comaprison to non smoker
- Non smoker give more tips in comaprison to smoker
- Customer who smoke gave highest tip

To distinguish the day

```python
sns.relplot(x= 'total_bill', y= 'tip', data= df, hue= 'day')
```

*Insights* : The day people visited most is Saturday

To distinguish between male and female

```python
sns.relplot(x= 'total_bill', y= 'tip', data= df, hue= 'sex')
```

## Line plot

relplot to make a line plot >> generating a data

To understanding the trending data

A time column/ sequences generally involved

```python
time = np.arange(500)
value = np.random.randn(500).cumsum()
```

```python
df = pd.DataFrame({'time': time, 'value': value})
df
```

```python
sns.relplot(x = 'time', y = 'value', kind = 'line', data = df, sort= True)
```

# Co-relation

How two features x and y varies.


If x/y are increasing/ decreasing or not

```python
df1.corr()
```

- If total bill increasing >> Tip will be increased by 67% and size is increased by 59%
- If tip is increasing >> total bill will be increased by 67% and size will be increased by 48%
- If size is increasing >> total bill will be increased by 59% and tip will be increased by 48%
