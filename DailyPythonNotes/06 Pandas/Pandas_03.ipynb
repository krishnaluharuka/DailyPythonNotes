## Statistical Information of Column

To get statistical information about numerical column of data frame , we use :

`dataframe.describe()`

```python
df.describe()
```

Data are of two types:
1. Qualitative ----> Categorical ----> String
2. Quantitative -----> Numerical

- Numerical are of two types:
- Countinuous : Float --> Any value on number line
- Discrete : Integer --> Only whole number, finite value.

- `count` : For each numerical column we have number of elements.
- `mean` : Avg of column
- `std` : standard deviation of the col
- `min` : Minimum value of any column
- `25%` : 25% percentile : Above than 25% of people
- `50%` : 50% percentile : Above than 50% of people
- `75%` : 75% percentile : Above than 75% of people
- `max` : Maximum value of column.

#### Five points summary:
- min
- 25%
- 50%
- 75%
- max

These five points give a lots of information about data.
